```python
!pip install transformers torch

import json
from transformers import BertTokenizer, BertModel
import torch

# JSON dosyasını oku
json_data = """
{
    "name": "Mehmet Aydın",
    "industry": "Manufacturing, Automotive, Robotics",
    "competence": [
        "Product Development",
        "Supply Chain Management",
        "Quality Control",
        "Operations Optimization"
    ],
    "focus": [
        "Robotics",
        "Automotive"
    ]
}
"""

# JSON verisini Python sözlüğüne dönüştür
data = json.loads(json_data)

# BERT tokenizer ve modelini yükle
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Her bir anahtarı ve değeri tokenleştir ve sayısal temsilleri al
for key, value in data.items():
    # Metni tokenleştir
    tokens = tokenizer(value, return_tensors='pt', padding=True, truncation=True)

    # Tokenleri modele giriş yap
    outputs = model(**tokens)

    # İlk gizli katmanın (embeddings) çıktılarını al
    embeddings = outputs.last_hidden_state

    # İlk tokenün sayısal temsilini yazdır
    print(f"Anahtar: {key}")
    print("Tokenler:")
    print(tokens)
    print("Sayısal Temsil:")
    print(embeddings[0])
    print("\n")
```

```python
import torch
from transformers import BertTokenizer, BertModel

# BERT modeli ve tokenizer'ını yükle
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# İki örnek JSON veri
cv_json = {
    "name": "Selma Yılmaz",
    "industry": "Fashion, Retail, E-commerce",
    "competence": [
        "Brand Management",
        "Market Research",
        "E-commerce Strategy",
        "Visual Merchandising"
    ],
    "focus": [
        "Fashion",
        "E-commerce"
    ]
}

startup_json = {
    "title": "Poland",
    "business_name": "PlayerGo",
    "category": "Poland",
    "based_in": "Krakow",
    "tags": "games, hardware, controllers, gaming, videogames",
    "founded": "2010"
}

# Her iki JSON verisi için işlemleri gerçekleştir
for data_name, data in [("CV", cv_json), ("Startup", startup_json)]:
    print(f"{data_name} JSON Verisi:")
    print(data)
    print("\n")

    # Her bir anahtar ve değer için işlem yap
    for key, value in data.items():
        print(f"Anahtar: {key}")
        print(f"Metin: {value}")

        # Metni tokenleştir
        tokens = tokenizer(value, return_tensors='pt', padding=True, truncation=True)
        print("Tokenler:")
        print(tokens)

        # Tokenlerin sayısal temsilini al
        with torch.no_grad():
            outputs = model(**tokens)
            last_hidden_state = outputs.last_hidden_state.mean(dim=1)
        print("Sayısal Temsil:")
        print(last_hidden_state)
        print("\n")

    print("---------------------------------------------------\n")
```
